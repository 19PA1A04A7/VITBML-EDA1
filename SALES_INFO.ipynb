```
import pandas as pd
from matplotlib import pyplot as plt
df=pd.read_csv('1000000 Sales Records.zip')
print(df)
```

```

cater_sale = df.groupby('Unit Price')['Units Sold'].mean()
plt.style.use('ggplot')
cater_sale.plot(kind = 'bar',)
plt.show()
```

```

cater_sale = df.groupby('Unit Price')['Units Sold'].mean()
plt.style.use('ggplot')
cater_sale.plot()
plt.show()
```

```
item_pro=df.groupby('Item Type')['Total Profit'].mean()
plt.style.use('ggplot')
item_pro.plot()
plt.show()
```

```
total_profit=df.groupby('Country')['Total Profit']
plt.style.use('ggplot')
total_profit.plot(kind='hist',)
plt.show()
```
